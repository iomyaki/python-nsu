Непосредственно задача:

**Первый пункт**

Необходимо реконструировать последовательность нуклеотидов от начала до конца (консенсус №1), ориентируясь на несколько условий:
1. Минимальное покрытие позиции - 80;
2. Если позиция отсутствует в таблице, то в последовательность ставится гэп "-";

Для данной последовательности необходимо:
1. Посчитать GC-состав;
2. Визуализировать соотношения А, T, G, C в виде bar-plot;
3. Оценить, насколько геном покрыт прочтениями;
4. Для этого построить график покрытия (можно использовать Kernel Density Estimation).





Как учесть пропущенные позиции? Для этого вам нужна информация про длину референсного генома.

Т.е. гэпы должны быть в местах, где покрытие меньше 80, и в местах, в которых отсутствует информация в таблице покрытия. Если таблица начинается с седьмой позиции , т.е. отсутствует информация о первых шести позициях (и в седьмой позиции наиболее часто встречается нуклеотид А), то последовательность будет следующей:



**Пункт два**

Необходимо реконструировать последовательность нуклеотидов от начала до конца (консенсус №2), ориентируясь на несколько условий:
1. Минимальное покрытие позиции - 80;
2. Если позиция отсутствует в таблице, то в последовательность ставится нуклеотид референсной последовательности;

Т.е. если в таблице нет информации о шестой позиции, то в консенсус будет поставлен шестой нуклеотид из референсной последовательности.

Для данной последовательности необходимо:
1. Посчитать GC-состав;
2. Использовать скользящее окно размером 1% от размера последовательности и шагом размера 10% от размера скользящего окна для подсчета GC-состава на каждой итерации. На основе полученных результатов визуализировать распределение GC-состава по полученной последовательности;
3. Аналогично сделать для АТ-состава.
4. Визуализировать два графика в одном фрейме.
